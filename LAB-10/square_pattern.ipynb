```python
from numba import cuda
import numpy as np
```

```python
@cuda.jit
def square_pattern(pattern):
    
    row = cuda.blockIdx.y * cuda.blockDim.y + cuda.threadIdx.y
    col = cuda.blockIdx.x * cuda.blockDim.x + cuda.threadIdx.x

    if row < pattern.shape[0] and col < pattern.shape[1]:
        pattern[row, col] = ord('*')
```

```python
N = 5      # Size of square

pattern = np.zeros((N, N), dtype=np.uint8)
```

```python
d_pattern = cuda.to_device(pattern)
```

```python
threads_per_block = (16, 16)

blocks_x = (N + threads_per_block[0] - 1) // threads_per_block[0]
blocks_y = (N + threads_per_block[1] - 1) // threads_per_block[1]

square_pattern[(blocks_x, blocks_y), threads_per_block](d_pattern)
```

```python
result = d_pattern.copy_to_host()
```

```python
print("Square Pattern:")

for i in range(N):
    print('   '.join(chr(x) for x in result[i]))
```
